```
import pandas as pd
import numpy as np
```

```
import keras
```

```
df = pd.read_csv('train.csv')
```

```
df.head()
```

```
df.shape
```

```
df.info()
```

```
t= df.iloc[:,1:].values
```

```
t.shape
```

```
t = t.reshape(-1,28,28,1)
```

```
t
```

```
y = df.iloc[:,0].values
```

```
y = keras.utils.to_categorical(y)
```

```
y.shape
```

```
from keras.layers import Dense
from keras.models import Sequential
from keras.layers import Convolution2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.layers import Dropout
```

```
model = Sequential()

model.add(Convolution2D(32,3,3,input_shape=(28,28,1),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))

model.add(Convolution2D(64,3,3,activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))

model.add(Convolution2D(64,3,3,activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))

model.add(Flatten())

model.add(Dense(20,activation='relu'))
model.add(Dense(20,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```
test_df = pd.read_csv('test.csv')
```

```
model.fit(t,y,epochs=30,validation_split=0.2,batch_size=128)
```

```
test_df.head()
```

```
x_test = test_df.iloc[:,:].values
```

```
x_test.shape
```

```
x_test = x_test.reshape(-1,28,28,1)
```

```
y_pred = model.predict(x_test)
```

```
y_pred
```

```
class_result=np.argmax(y_pred,axis=-1)
```

```
class_result.shape
```

```
image_id=np.array(range(1,28001))
```

```
ImageId = pd.Series(image_id)
```

```
Label = pd.Series(class_result)
```

```
submission = pd.DataFrame()
```

```
submission['ImageId']=ImageId
```

```
submission['Label']=Label
```

```
submission.to_csv('submission.csv',index=False)
```
